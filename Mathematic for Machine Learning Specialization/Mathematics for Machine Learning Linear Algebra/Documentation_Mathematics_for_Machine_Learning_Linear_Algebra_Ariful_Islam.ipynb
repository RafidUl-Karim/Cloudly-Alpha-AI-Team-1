## **Properties of Dot Product**

The **dot product** of two vectors **a** and **b** has the following key properties:

### **1. Commutative Property**
a · b = b · a  
- The order of multiplication does not change the result.

### **2. Distributive Property**
a · (b + c) = a · b + a · c  
- Dot product distributes over vector addition.

### **3. Scalar Multiplication Property**
(c * a) · b = c * (a · b)  
- Multiplying a vector by a scalar before taking the dot product is the same as multiplying the dot product by the scalar.

### **4. Zero Vector Property**
a · 0 = 0  
- The dot product of any vector with the zero vector is always zero.

### **5. Orthogonality Property**
a · b = 0  if a is perpendicular to b  
- If two vectors are perpendicular (90° angle), their dot product is zero.

### **6. Magnitude and Angle Relationship**
a · b = |a| * |b| * cos(θ)  
- The dot product relates to the magnitudes of the vectors and the cosine of the angle between them.

These properties make the dot product useful in **machine learning, physics, and vector analysis**! 🚀

##Matrix Transformation

**📌 Matrix Transformations in Python**

This notebook demonstrates different **matrix transformations** using NumPy.  
Each transformation is applied to a **vector (2,3)** to observe the effects.

**🔹 Transformations Covered:**
1. **Swap x and y** → Reflects over the line y = x  
2. **Reflection over X-axis** → Flips the y-coordinate  
3. **Reflection over Y-axis** → Flips the x-coordinate  
4. **90-degree Counterclockwise Rotation** → Rotates left  
5. **180-degree Rotation** → Rotates completely  
6. **Shear in X-direction** → Skews along the x-axis  
7. **Shear in Y-direction** → Skews along the y-axis  


image.png

```python
import numpy as np

# Function to apply a transformation matrix
def transform(matrix, vector):
    return np.dot(matrix, vector)

# Example matrices
swap_xy = np.array([[0, 1], [1, 0]])  # Swap x and y (Reflection over y=x)
reflect_x = np.array([[1, 0], [0, -1]])  # Reflect over X-axis
reflect_y = np.array([[-1, 0], [0, 1]])  # Reflect over Y-axis
rotate_90 = np.array([[0, -1], [1, 0]])  # 90-degree CCW Rotation
rotate_180 = np.array([[-1, 0], [0, -1]])  # 180-degree Rotation
shear_x = np.array([[1, 2], [0, 1]])  # Shear in X-direction (k=2)
shear_y = np.array([[1, 0], [2, 1]])  # Shear in Y-direction (k=2)

# Example vector
vector = np.array([2, 3])

# Apply transformations
print("Original Vector:", vector)
print("Swap x and y:", transform(swap_xy, vector))
print("Reflection over X-axis:", transform(reflect_x, vector))
print("Reflection over Y-axis:", transform(reflect_y, vector))
print("90-degree CCW Rotation:", transform(rotate_90, vector))
print("180-degree Rotation:", transform(rotate_180, vector))
print("Shear in X-direction:", transform(shear_x, vector))
print("Shear in Y-direction:", transform(shear_y, vector))
```

## Orthonormal Bases & Orthogonal Matrices
* A transpose of a matrix is created by flipping rows and columns.

* If a square matrix’s columns are orthonormal (unit length + perpendicular), it’s called an orthogonal matrix.

For an orthogonal matrix

* This makes the inverse easy to compute.

* The determinant of an orthogonal matrix is either +1 or −1:

  * +1 means regular orientation

  * −1 means it flips space (reflection)

* Dot products with orthonormal basis vectors give projections, simplifying coordinate transformations.

* Using orthonormal bases is highly beneficial in data science (e.g., PCA) because transformations:

  * Are reversible

  * Preserve distances and angles

  * Are computationally simple
